```python
!pip install word2number
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model as lm
from word2number import w2n
```

```python
df = pd.read_csv('2) hiring.csv')
df
```

```python
i = 0
while i < 8:
  if type(df.experience[i]) is str:
    df.experience[i] = w2n.word_to_num(df.experience[i])
  i+=1
df[['experience']] = df[['experience']].fillna(df[['experience']].tail(len(df)-2).median())
df[['test_score(out of 10)']] = df[['test_score(out of 10)']].fillna(df[['test_score(out of 10)']].median())
df
```

```python
reg = lm.LinearRegression()
reg.fit(df[['experience','test_score(out of 10)','interview_score(out of 10)']],df[['salary($)']])
```

```python
df.loc[len(df)] = [2,9,6,int(reg.predict([[2,9,6]]).squeeze().round())]
df.loc[len(df)] = [12,10,10,int(reg.predict([[12,10,10]]).squeeze().round())]
df
```

```python
df.to_csv('updatedhiring.csv', index=True)
```
