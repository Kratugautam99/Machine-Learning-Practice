```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model as  lm
```

```python
df = pd.read_csv('1) canada_per_capita_income.csv')
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df[['year']], df[['per capita income (US$)']], color='purple', marker='1')
```

```python
reg = lm.LinearRegression()
reg.fit(df[['year']],df[['per capita income (US$)']])
```

```python
i = 2017
while not(i==2021):
    df.loc[len(df.index)] = [i,reg.predict([[i]]).squeeze().round(5)]
    i+=1
```
