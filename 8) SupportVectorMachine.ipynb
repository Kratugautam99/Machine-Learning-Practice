```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
import pandas as pd
```

```python
digits = load_digits()
inputs = digits.data
targets = digits.target
x_train, x_test, y_train, y_test = train_test_split(inputs, targets, test_size=0.3, random_state=10)
```

```python
model = SVC()
print('That\'s the max accuracy you can achieve')
model.fit(inputs,targets)
model.score(inputs,targets)
```

```python
#Default at this point gamma = 'scale', C = 1.0 , kernel = 'rbf'
model = SVC()
model.fit(x_train, y_train)
model.score(x_test, y_test)
```

```python
#Tuning Gamma
model = SVC(gamma='auto')
model.fit(x_train, y_train)
model.score(x_test, y_test)
#It does not worked it only reduced to 0.094.. at every number. also 'auto' is also doing poor here but better than any manual input.
```

```python
#Tuning Regulariztion
model = SVC(C=100)
model.fit(x_train, y_train)
model.score(x_test, y_test)
#It increased the accuracy by very small amount.
```

```python
#Tuning Kernel
model = SVC(kernel='poly')
model.fit(x_train, y_train)
model.score(x_test, y_test)
#It decreased the accuracy by very small amount in both linear and poly also in both answer is equal.
```
