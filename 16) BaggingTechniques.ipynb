```python
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from scipy.stats import zscore
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import cross_val_score
```

```python
df = pd.read_csv('7) heart.csv')
df1 = df[['Age','RestingBP','Cholesterol','FastingBS','MaxHR','HeartDisease','Oldpeak']]
z_scores = np.abs(zscore(df1))
df1 = df[(z_scores < 3).all(axis=1)]
df1
```

```python
le = LabelEncoder()
df1 = df1.apply(le.fit_transform)
ct = ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[1,2,6,8,10])],remainder='passthrough')
x = df1.drop('HeartDisease',axis=1)
y = df1['HeartDisease']
x = ct.fit_transform(x)
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
svm = SVC()
kfoldsvm = cross_val_score(svm,x,y,cv=123)
bagging_modelsvm = BaggingClassifier(estimator=SVC(), n_estimators=10, random_state=0)
bagging_modelsvm.fit(x_train, y_train)
svm.fit(x_train,y_train)
print(f'''bagging_modelsvm:
{bagging_modelsvm.score(x_test,y_test)},
kfoldsvm:
{kfoldsvm.mean()},
regularsvm:
{svm.score(x_test,y_test)}''')
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
dt = DecisionTreeClassifier()
kfolddt = cross_val_score(dt,x,y,cv=123)
bagging_modeldt = BaggingClassifier(estimator=DecisionTreeClassifier(), n_estimators=10, random_state=0)
bagging_modeldt.fit(x_train, y_train)
dt.fit(x_train,y_train)
print(f'''bagging_modeldt:
{bagging_modeldt.score(x_test,y_test)},
kfolddt:
{kfolddt.mean()},
regulardt:
{dt.score(x_test,y_test)}''')
```
