```python
#Part 1
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('4) carprices.csv')
df
```

```python
dummies = pd.get_dummies(df[['Car Model']])
dummies
```

```python
merged = pd.concat([df,dummies],axis = 'columns').drop(['Car Model'], axis = 'columns')
```

```python
merged = merged.iloc[:,:-1]
merged
```

```python
from sklearn.linear_model import LinearRegression
model1 = LinearRegression()
X_train1 = np.array(merged.drop(['Sell Price($)'], axis='columns'))
y_train1 = merged['Sell Price($)']
model1.fit(X_train1, y_train1)
```

```python
#Part 2
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
dfle = df
le = LabelEncoder()
ohe = OneHotEncoder()
dfle['Car Model'] = le.fit_transform(dfle['Car Model'])
print(dfle)
```

```python
x_train2 = dfle[['Mileage','Age(yrs)','Car Model']].values
ct = ColumnTransformer([('Car Model', OneHotEncoder(), [2])], remainder = 'passthrough')
x_train2 = ct.fit_transform(x_train2)
x_train21 = np.hstack((x_train2[:,3:5], x_train2[:,:2]))
x_train2 = x_train21
y_train2 = dfle['Sell Price($)'].values
model2 = LinearRegression()
model2.fit(x_train2, y_train2)
x_train2
```

```python
#As Our both models are built, we can use and find solutions to our exercise.
#1
model1.predict([[45000,4,0,0]])
```
